```python
def get_op_code(tp, month, strike):
    file_path = 'E:\\文件管理集合\\代码\\Python\\PyQt5快速开发与实战\\Option\\contract_info.csv'
    codes_df = pd.read_csv(file_path)
    temp_df = codes_df[(codes_df['type']==tp)&(codes_df['month']==month)&(codes_df['strike']==strike)]
    print(temp_df.shape)
    if temp_df.shape[0] == 1:
        return temp_df.iat[0,0]
    
get_op_code('call', 1906, 2400)
```
