- **前馈运算**（feed-forward）：卷积神经网络通过卷积（convolution）操作、汇合（pooling）操作和非线性激活函数（non-linear activation function）映射等一系列操作的层层堆叠，将高层语义信息逐层由原始数据输入层中抽取出来，逐层抽象。

- 通过计算预测值与真实值之间的误差或损失（loss），凭借反向传播算法（back-propagation algorithm）将误差或损失由最后一层逐层向前反馈（back-forward），更新每层参数，并在更新参数后再次前馈，如此往复，直到网络模型收敛，从而达到项目训练的目的。

最后，整个网络以损失函数的计算结束。若$y$是输入$x^1$对应的真实标记，则损失函数为：
$$z=L(x^L,y)$$
以回归问题为例，常用的$l_2$损失函数即可作为卷积神经网络的目标函数，此时有$$z=L_{regression}(x^L,y)=\frac{1}{2}||x^L-y||^2$$
若对于分类问题，网络的目标函数常采用交叉熵（cross entropy）损失函数，有$$z=L_{regression}(x^L,y)=-\sum_iy_ilog(p_i)$$

在利用交叉熵损失函数训练后得到的网络中，$X^L$的每一维可表示$x^1$分别隶属于$C$个类别的后验概率。其对应的预测标记为：$$\underset{i}{arg\ max}\ x_i^L$$

_____
深度学习模型采用随机梯度下降法（Stochastic Gradient Descent）和误差反向传播（error back propogation）进行模型参数更新。

1. **批处理的随机梯度下降法**：在训练模型阶段随机选取$n$个样本作为一批（batch）样本，先通过前馈运算得到预测并计算其误差，后通过梯度下降法更新参数，梯度从后往前逐层反馈，直至更新到网络的第一层参数，这样的一个参数更新过程称为一个“批处理过程”（mini-batch）。
2. 不同批处理之间按照无放回抽样遍历所有训练集样本，遍历一次训练样本称为“一轮”（epoch）。
3. 批处理样本的大小（batch size）不宜设置过小，过小时，由于样本采样随机，按照该样本上的误差更新模型参数不一定在全局上最优，会使得训练过程产生震荡。而批处理的上限主要取决于硬件资源的限制，一般设32、64、128或256即可。


- 前馈运算：对样本进行推理和预测。
- 反馈运算：将预测误差反向传播逐层更新参数。
- 上述两种运算依次交替迭代完成模型的训练过程。

#### 反向传播算法

- 式1.7，第$i$层误差对该层参数的导数：$$\frac{\partial z}{\partial (vec(x^{i+1}))^T}=\frac{\partial z}{\partial (vec(x^{i+1})^T)}\cdot \frac{\partial vec(x^{i+1})}{\partial (w^i)^T}$$
- 式1.8，第$i$层误差对该层输入数据的导数：$$\frac{\partial z}{\partial vec(x^i)^T}=\frac{\partial z}{\partial (vec(x^{i+1})^T)}\cdot \frac{\partial vec(x^{i+1})}{\partial (x^i)^T}$$
- 式1.6，参数更新：$$w^i\leftarrow w^i-\eta\frac{\partial z}{\partial w^i}$$
