### 感知机
- 物理意义：将输入空间（特征空间）划分为正负两类的分离超平面。
- 定义函数为：$$f(\vec{x})=sign(\vec{w}\cdot \vec{x} +b)$$
- 任取一个误分类的样本点$(\vec{x}_i, y_i)$，则$\vec{x}_i$距离超平面的距离为：$$\frac{1}{||\vec{w}||_2}|\vec{w}\cdot \vec{x}_i+b|=\frac{-y_i(\vec{w}\cdot \vec{x}_i+b)}{||\vec{w}||_2}$$
- 损失函数为：误分类点到超平面$S$的总距离。最终形式为：$$L(\vec{w},b)=-\sum_{\vec{x}_i\in M}y_i(\vec{w}\cdot \vec{x}_i+b)$$
其中，$M$为误分类点的集合。

当一个实例点被误分类时，调整$\vec{w}$,$b$使得分离平面向该误分类点的一侧移动，以减少该误分类点与超平面间的距离，直至超平面对所有的实例都能够正确分类为止。其原始形式获得的解并不是唯一的。

原始形式：
> 1. 选取初始值$\vec{w}_0$，$b_0$
> 2. 在训练数据集中选取数据$(\vec{x}_i,y_i)$
> 3. 若$y_i(\vec{w}\cdot \vec{x}_i+b)\leq 0$（即该实例为误分类点）则更新参数：$$\begin{matrix}
\vec{w}\leftarrow \vec{w}+\eta y_i\vec{x}_i\\ 
b\leftarrow b+\eta y_i
\end{matrix}$$

如果考虑去初始值$\vec{w}_0=\vec{0}$，$b_0=0$，则迭代公式可改为：$$\begin{matrix}
\vec{w}=\sum_{i=1}^N\alpha_i y_i \vec{x}_i\\ 
b=\sum_{i=1}^N\alpha_i y_i
\end{matrix}$$
对偶形式为：
> 1. $\vec{\alpha}\leftarrow \vec{0}$，$b\leftarrow 0$
> 2. 在训练数据集中选取数据$(\vec{x}_i,y_i)$
> 3. 若$y_i\sum_{j=1}^N\alpha_jy_j\vec{x}_j\cdot \vec{x}_i+b\leq 0$，则更新参数。

____
标准BP算法的原理，得到误差逆传播算法：
输出：
> 1. 输入层到隐含层的$n\times q$个权值$v_{ih}$，$i=1,2,\cdots,n$；$h=1,2,\cdots,q$
> 2. 隐含层到输出层的$q \times m$个权值$w_{hj}$，$h=1,2,\cdots,q$；$j=1,2,\cdots,m$
> 3. $q$个隐含层神经元的阈值$\gamma_h$，$h=1,2,\cdots,q$
> 4. $m$个输出层神经元的阈值$\theta_j$；$j=1,2,\cdots,m$

算法步骤：
> 1. 在0,1范围内随机初始化网络中所有的连接权值和阈值。
> 2. 对训练数据集$T$中的样本点迭代，直到到达停止条件为止，迭代过程如下：
    - 取样本点$(\vec{x}_k,\vec{y}_k)$，根据当前参数计算该样本的输出$\vec{\hat{y}}_k=(\hat{y}_k^{(1)},\hat{y}_k^{(2)},\cdots,\hat{y}_k^{(m)})^T$
    - 根据$g_j=-\hat{y}_k^{(j)}(1-\hat{y}_k^{(j)})(\hat{y}_k^{(j)}-y_k^{(j)})$，计算输出层神经元的梯度项$g_j$，其中$\vec{y}_k=(y_k^{(1)},y_k^{(2)},\cdots,y_k^{(m)})^T$。
    - 根据$e_h=b_h(1-b_h)\sum_{j=1}^mg_jw_{hj}$计算隐层神经元的梯度项$e_h$。
    - 根据下列式子更新连接权值$w_{hj}$，$v_{ih}$和阈值$\theta_j$，$\gamma_h$：$$\begin{matrix}
\Delta w_{hj}=\eta g_jb_h\\ 
\Delta \theta_j = -\eta g_j\\ 
\Delta v_{ih}=\eta e_hx^{(i)}\\ 
\Delta \gamma_h=-\eta e_h
\end{matrix}$$

标准BP算法每次更新只针对单个样例，参数更新非常频繁，而且对于不同的样例进行更新的效果可能出现“抵消”现象。

BP神经网络缓解过拟合的策略：
1. “早停”策略：将数据集分成训练数据集和验证数据集两类，如果训练集误差降低但是验证集误差升高，则停止训练。同时返回具有最小验证集误差的连接权值和阈值。
2. “正则化”策略：$E=\frac{1}{N}\sum_{k=1}^NE_k+\lambda\sum_iw_i^2$，其中$\lambda\sum_iw_i^2$刻画了网络复杂度。

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
```

```python
def create_data(n):
    np.random.seed(1)
    x_11 = np.random.randint(0, 100, (n, 1))
    x_12 = np.random.randint(0, 100, (n, 1))
    x_13 = 20 + np.random.randint(0, 10, (n, 1,))
    x_21 = np.random.randint(0, 100, (n, 1))
    x_22 = np.random.randint(0, 100, (n, 1))
    x_23 = 10 - np.random.randint(0, 10, (n, 1,))
    
    new_x_12 = x_12 * np.sqrt(2) / 2 - x_13 * np.sqrt(2) / 2
    new_x_13 = x_12 * np.sqrt(2) / 2 + x_13 * np.sqrt(2) / 2
    new_x_22 = x_22 * np.sqrt(2) / 2 - x_23 * np.sqrt(2) / 2
    new_x_23 = x_22 * np.sqrt(2) / 2 + x_23 * np.sqrt(2) / 2
    
    plus_samples = np.hstack([x_11, new_x_12, new_x_13, np.ones((n, 1))])
    minus_samples = np.hstack([x_21, new_x_22, new_x_23, -np.ones((n, 1))])
    samples = np.vstack([plus_samples, minus_samples])
    np.random.shuffle(samples)  # 混洗数据
    return samples
```

```python
def plot_samples(ax, samples):
    Y = samples[:, -1]
    position_p = Y == 1
    position_m = Y == -1
    ax.scatter(samples[position_p, 0], samples[position_p, 1], samples[position_p, 2], marker='+', color='b')
    ax.scatter(samples[position_m, 0], samples[position_m, 1], samples[position_m, 2], marker='^', color='y')
    
fig = plt.figure()
ax = Axes3D(fig)
data = create_data(100)
plot_samples(ax, data)
ax.legend(loc='best')
plt.show()
```

```python
def perceptron(train_data, eta, w_0, b_0):
    x = train_data[:, :-1]
    y = train_data[:, -1]
    length = train_data.shape[0]
    w = w_0
    b = b_0
    step_num = 0
    while True:
        i = 0
        while i < length:
            step_num += 1
            x_i = x[i].reshape((x.shape[1], 1))
            y_i = y[i]
            if y_i * (np.dot(np.transpose(w), x_i) + b) <= 0:
                w = w + eta * y_i * x_i
                b = b + eta * y_i
                break
            else:
                i = i + 1
        if i == length:
            break
    return (w, b, step_num)
```

```python
def create_hyperplane(x, y, w, b):
    return (-w[0][0] *x - w[1][0] * y - b) / w[2][0]
```

```python
data = create_data(100)
eta, w_0, b_0 = 0.1, np.ones((3, 1), dtype=float), 1
w, b, num = perceptron(data, eta, w_0, b_0)

fig = plt.figure()
plt.suptitle("perceptron")
ax = Axes3D(fig)

plot_samples(ax, data)

x = np.linspace(-30, 100, 100)
y = np.linspace(-30, 100, 100)
x, y = np.meshgrid(x, y)
z = create_hyperplane(x, y, w, b)
ax.plot_surface(x, y, z,rstride=1, cstride=1, color='g', alpha=0.2)

ax.legend(loc='best')
plt.show()
```

学习速度$\eta$的选择：
> 1. 如果$\eta$较小，则达到收敛所需要迭代的次数就会非常高
> 2. 如果$\eta$较大，则每次迭代可能不会减小代价函数的结果，甚至会超过局部最小值导致无法收敛。
> 3. 通常的经验是，从以下几个数值开始试验$\eta$的值：0.001，0.003，0.01，0.03，0.1，0.3，1

#### 多层感知机算法MLPClassifier
**sklearn.neural_network.MLPClassifier(hidden_layer_sizes=(100, ), activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate='constant', learning_rate_init=0.001, power_t=0.5, max_iter=200, shuffle=True, random_state=None, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True, early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08, n_iter_no_change=10)**
> 1. hidden_layer_sizes：元组，指定了隐含层的结构。
> 2. activation：字符串，指定了激活函数的类型。
    - logistic
    - tanh
    - relu
> 3. solver：字符串，指定采用的最优化算法的类型。
    - lbfgs：伪牛顿算法
    - sgd：随机梯度下降法
    - adam：stochastic gradient-based最优化算法
> 4. alpha：浮点数，给出正则化项的系数。
> 5. learning_rate_init：初始学习率。使用sgd、adam算法时才有意义。
> 6. early_stopping：布尔值，若为True，则启用早停策略。
> 7. validation_fraction：位于0和1之间的浮点数。划分训练集的该份额用于验证集。

属性：
> 1. classes_：每个输出的类别
> 2. loss_：当前损失函数值
> 3. coefs_：给出权值
> 4. intercepts_：给出阈值
> 5. n_iter：实际迭代次数

如何设置隐含层神经元的个数理论上尚未解决，实际中靠“试错法”来调整。

```python
def create_data_no_linear_2d(n):
    np.random.seed(1)
    x_11 = np.random.randint(0, 100, (n, 1))
    x_12 = 10 + np.random.randint(-5, 5, (n, 1))
    x_21 = np.random.randint(0, 100, (n, 1))
    x_22 = 20 + np.random.randint(0, 10, (n, 1,))
    x_31 = np.random.randint(0, 100, (int(n/10), 1))
    x_32 = 20 + np.random.randint(0, 10, (int(n/10), 1))
    
    new_x_11 = x_11 * np.sqrt(2) / 2 - x_12 * np.sqrt(2) / 2
    new_x_12 = x_11 * np.sqrt(2) / 2 + x_12 * np.sqrt(2) / 2
    new_x_21 = x_21 * np.sqrt(2) / 2 - x_22 * np.sqrt(2) / 2
    new_x_22 = x_21 * np.sqrt(2) / 2 + x_22 * np.sqrt(2) / 2
    new_x_31 = x_31 * np.sqrt(2) / 2 - x_32 * np.sqrt(2) / 2
    new_x_32 = x_31 * np.sqrt(2) / 2 + x_32 * np.sqrt(2) / 2
    
    plus_samples = np.hstack([new_x_11, new_x_12, np.ones((n, 1))])
    minus_samples = np.hstack([new_x_21, new_x_22, -np.ones((n, 1))])
    err_samples = np.hstack([new_x_31, new_x_32, np.ones((int(n/10), 1))])
    samples = np.vstack([plus_samples, minus_samples, err_samples])
    np.random.shuffle(samples)  # 混洗数据
    return samples
```

```python
def plot_samples_2d(ax, samples):
    Y = samples[:, -1]
    position_p = Y == 1
    position_m = Y == -1
    ax.scatter(samples[position_p, 0], samples[position_p, 1], marker='+', label='+', color='b')
    ax.scatter(samples[position_m, 0], samples[position_m, 1], marker='^', label='-', color='y')
    
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
data = create_data_no_linear_2d(100)
plot_samples_2d(ax, data)
ax.legend(loc='best')
plt.show()
```

```python
def predict_with_MLPClassifier(ax, train_data):
    train_x = train_data[:, :-1]
    train_y = train_data[:, -1]
    clf = MLPClassifier(activation='logistic', max_iter=1000)
    clf.fit(train_x, train_y)
    print(clf.score(train_x, train_y))
    
    x_min, x_max = train_x[:, 0].min() - 1, train_x[:, 0].max() + 2
    y_min, y_max = train_x[:, 1].min() - 1, train_x[:, 1].max() + 2
    plot_step = 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
data = create_data_no_linear_2d(500)
predict_with_MLPClassifier(ax, data)
ax.legend(loc='best')
plt.show()
```
